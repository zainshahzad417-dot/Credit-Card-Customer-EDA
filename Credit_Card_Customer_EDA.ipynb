```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Credit Card Customer Data.csv')
corr=df[['Total_Credit_Cards','Total_visits_bank','Total_visits_online','Total_calls_made']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('All columns co-relation')
plt.show()

corr=df[['Customer Key','Avg_Credit_Limit']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('All columns co-relation')
plt.show()

corr=df[['Customer Key','Avg_Credit_Limit','Total_Credit_Cards','Total_visits_bank','Total_visits_online','Total_calls_made']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('All columns co-relation')
plt.show()
```

```python
import numpy as np
import pandas as pd
df = pd.read_csv('Credit Card Customer Data.csv')

numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

category_columns = df.select_dtypes(include=['object']).columns
for col in category_columns:
    df[col] = df[col].fillna(df[col].mode()[0])
threshold=0.5
threshold_value = len(df) * threshold
df.dropna(thresh=threshold_value, axis=1, inplace=True)

cleaned_df = df.copy()
cleaned_df.to_csv('cleaned_Credit Card Customer Data.csv', index=False)
```
